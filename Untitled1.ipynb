```python
import well_log_analysis.dataset as data
import well_log_analysis.models.mlp as mlp
import lasio
import torch
from torch.utils.data import DataLoader
import numpy as np
```

```python
sequence_length=32
batch_size=128
epochs=256
```

```python
dataset = data.WellDataset('data/dataset1/J10035.las', ['GR', 'DEN', 'RT', 'DTCO', 'DTSM'], ['GR', 'DEN', 'RT', 'DTCO', 'DTSM'], sequence_length=sequence_length)
#dataset = data.WellDataset('data/dataset1/J10025.las', ['GR', 'DEN', 'RT', 'DTCO', 'DTSM'], ['GR', 'DEN', 'RT', 'DTCO', 'DTSM'], sequence_length=sequence_length)
#dataset = dataset1 + dataset2
#print(dataset)
```

```python
model = mlp.MLP(input_features=3, output_features=2, hidden_layers=[128, 64, 32], sequence_length=sequence_length).cuda()
```

```python
loader = DataLoader(dataset, batch_size = batch_size, drop_last=True, shuffle=True)
```

```python
dataset_mean = dataset.mean()
dataset_std = dataset.std()
```

```python
loss_fn = torch.nn.SmoothL1Loss()
optim = torch.optim.Adam(model.parameters(), weight_decay=1e-4)
```

```python
for epoch in range(epochs):
    losses = []
    for batch in loader:
        optim.zero_grad()
        batch = ((batch - dataset_mean) / dataset_std).cuda()
        x = batch[:, :, 0:3].float()
        y = batch[:, sequence_length // 2, 3:5].float()
        pred_y = model(x)
        loss = loss_fn(pred_y, y)
        loss.backward()
        optim.step()
        losses.append(loss.cpu().detach().clone())
    print(np.mean(losses))
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib ipympl
model.eval()
names = ['data/dataset1/J10039.las', 'data/dataset1/J10038.las', 'data/dataset1/J10025.las']
fig, axs = plt.subplots(len(names), 2, figsize=(14, 8*len(names)), sharex=True, dpi=80)
fig.suptitle("Sequenced MLP Performance on 3 unseen wells.")
for i in range(len(names)):
    test_dataset = data.WellDataset(names[i], ['GR', 'DEN', 'RT', 'DTCO', 'DTSM'], ['GR', 'DEN', 'RT', 'DTCO', 'DTSM'], sequence_length=sequence_length)
    pred_dtco = []
    dtco = []
    pred_dtsm = []
    dtsm = []

    for instance in test_dataset:
        instance = (instance - dataset_mean) / dataset_std

        pred_y = model(torch.from_numpy(instance[:, 0:3]).unsqueeze(0).cuda().float())
        y = torch.from_numpy(instance[sequence_length // 2, 3:5]).unsqueeze(0)
        pred_dtco.append(pred_y[0][0].cpu().detach())
        dtco.append(y[0][0])

        pred_dtsm.append(pred_y[0][1].cpu().detach())    
        dtsm.append(y[0][1])
    axs[i][0].plot(dtsm, range(len(dtsm)), color='blue', linewidth=0.5, label="Ground Truth DTSM")
    axs[i][0].plot(pred_dtsm, range(len(dtsm)), color='red', linewidth=0.5, label="Predicted DTSM")
    axs[i][0].legend()
    axs[i][0].set_title("DTSM for Well {}".format(i))
    axs[i][1].plot(dtco, range(len(dtco)), color='blue', linewidth=0.5, label="Ground Truth DTCO")
    axs[i][1].plot(pred_dtco, range(len(dtco)), color='red', linewidth=0.5, label = "Predicted DTCO")
    axs[i][1].legend()
    axs[i][1].set_title("DTCO for Well {}".format(i))
    correlation_matrix = np.corrcoef(dtsm, pred_dtsm)
    correlation_xy = correlation_matrix[0,1]
    r_squared = correlation_xy**2

    print("R2 Score of DTSM for Well {}: {}".format(i, r_squared))
    correlation_matrix = np.corrcoef(dtco, pred_dtco)
    correlation_xy = correlation_matrix[0,1]
    r_squared = correlation_xy**2
    print("R2 Score of DTCO for Well {}: {}".format(i, r_squared))
    print("RMSE of DTSM for Well {}: {}".format(i, np.sqrt(np.mean(((np.array(dtsm) - np.array(pred_dtsm)) ** 2)))))
    print("RMSE of DTCO for Well {}: {}".format(i, np.sqrt(np.mean(((np.array(dtco) - np.array(pred_dtco)) ** 2)))))
    
```
